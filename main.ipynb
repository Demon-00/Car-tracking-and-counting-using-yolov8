```python
import cv2
import pandas as pd
import numpy as np
from ultralytics import YOLO
from tracker import Tracker

model = YOLO('yolov8s.pt')

# to print the mouse pointer location
def RGB(event, x, y, flags, param):
    if event == cv2.EVENT_MOUSEMOVE:
        colorsBGR = [x, y]
        print(colorsBGR)

# cap = cv2.VideoCapture('ovehicle-counting.mp4')
cap = cv2.VideoCapture('highway_-_20090 (1080p).mp4')

tracker = Tracker()
count = 0
# cy1 = 422
# cy2 = 434
# cy3 = 446
# offset = 6
cy1 = 578
cy2 = 598
cy3 = 618
offset = 8
vh_down = {}
downcount = 0
vh_up = {}
upcount = 0

while True:
    (grabbed, frame) = cap.read()
    if not grabbed:
        break
    count += 1
    if count % 3 != 0:
        continue
    frame = cv2.resize(frame, (1080, 720))

    results = model.predict(frame)
    a = results[0].boxes.data
    px = pd.DataFrame(a).astype("float")
    list = []

    for index, row in px.iterrows():
        x1 = int(row[0])
        y1 = int(row[1])
        x2 = int(row[2])
        y2 = int(row[3])
        d = int(row[5])
        list.append([x1, y1, x2, y2])
    bbox_id = tracker.update(list)
    for bbox in bbox_id:
        x3, y3, x4, y4, id = bbox
        cx = int(x3 + x4) // 2
        cy = int(y3 + y4) // 2
        # exit
        if cy3 < (cy + offset) and cy3 > (cy - offset):
            vh_up[id] = cy
        if id in vh_down:
            if cy2 > (cy + offset) and cy2 < (cy - offset):
                cv2.circle(frame, (cx, cy), 4, (0, 0, 255), -1)
                cv2.putText(frame, str(id), (cx, cy), cv2.FONT_HERSHEY_COMPLEX, 0.8, (0, 255, 255), 1)
                upcount = upcount + 1
                # bbox_id = tracker.unregister(id)
        # enter
        if cy1 < (cy + offset) and cy1 > (cy - offset):
            vh_down[id] = cy
        if id in vh_down:
            if cy2 < (cy + offset) and cy2 > (cy - offset):
                cv2.circle(frame, (cx, cy), 4, (0, 0, 255), -1)
                cv2.putText(frame, str(id), (cx, cy), cv2.FONT_HERSHEY_COMPLEX, 0.8, (0, 255, 255), 1)
                downcount = downcount + 1
                # bbox_id = tracker.unregister(id)
    # # line 1 top
    # cv2.line(frame, (179, 422), (936, 422), (255, 255, 255), 1)
    # # line 2 mid
    # cv2.line(frame, (173, 434), (942, 434), (255, 255, 255), 1)
    # # line3 bot
    # cv2.line(frame, (167, 446), (948, 446), (255, 255, 255), 1)
    # line 1 top
    cv2.line(frame, (23, 578), (1056, 578), (255, 255, 255), 1)
    # line 2 mid
    cv2.line(frame, (23, 598), (1056, 598), (255, 255, 255), 1)
    # line3 bot
    cv2.line(frame, (23, 618), (1056, 618), (255, 255, 255), 1)
    cv2.putText(frame, ('Enter : ')+str(downcount), (60, 40), cv2.FONT_HERSHEY_COMPLEX, 0.8, (255, 255, 255), 1)
    cv2.putText(frame, ('Exit : ')+str(upcount), (60, 60), cv2.FONT_HERSHEY_COMPLEX, 0.8, (255, 255, 255), 1)
    cv2.imshow('RGB', frame)
    cv2.setMouseCallback('RGB', RGB)
    key = cv2.waitKey(1)

    # if you press 0, it will stop the program because waitKey also captures the key that was pressed on the keyboard
    if key == 0:
        break

# cap.release()
cv2.destroyAllWindows()
```
